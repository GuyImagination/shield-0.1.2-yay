## GPS Accuracy Using Root Mean Squared Error (RMSE)

```python
from sklearn.metrics import mean_squared_error
import numpy as np
import pandas as pd
import math

data = pd.read_excel('gps_data.xlsx')
lat = data['Latitude']
lon = data['Longitude']
true_lat = data['True Latitude']
true_lon = data['True Longitude']
true_dist1, true_dist2 = [], []
dist1, dist2 = [], []
mse1, mse2 = [], []
loc_rmse = []

dlat = np.deg2rad(lat - true_lat)
dlon = np.deg2rad(lon - true_lon)
lat = np.deg2rad(lat)
lon = np.deg2rad(lon)
true_lat = np.deg2rad(true_lat)
true_lon = np.deg2rad(true_lon)

for i in lat:
    true_dist_value = 0
    true_dist.append(true_dist_value)

# Haversine Formula
rEarth = 6371000
a = (np.sin(dlat / 2) ** 2) + np.cos(true_lat) * np.cos(lat) * ((np.sin(dlon / 2) / 2) ** 2)
c = 2 * np.arcsin(np.sqrt(a))
dist = rEarth * c
```

### First Location (JP Rizal Street, Brgy, Pililla - Jala-Jala - Pakil Rd, Pililla, 1910 Rizal)

```python
data[0:30]
```

### First Location RMSE

```python
dist1 = dist[0:30]
true_dist1 = true_dist[0:30]
mse1 = mean_squared_error(dist1, true_dist1)
np.sqrt(mse1)
print(np.sqrt(mse1))
loc_rmse.append(np.sqrt(mse1))
```

### Second Location (147 C. Tolentino Compound Banaag Street Pineda Pasig City)

```python
data[30:60]
```

```python
dist2 = dist[30:60]
true_dist2 = true_dist[30:60]
mse2 = mean_squared_error(dist2, true_dist2)
np.sqrt(mse2)
print(np.sqrt(mse2))
loc_rmse.append(np.sqrt(mse2))
```

## Remarks

The RMSE serves as a guide to determine the accuracy of a GPS within a specific range. It tells the observer that their true location might be within *n* meters of the true location (*n* is the RMSE).

Based on the data gathered by the researchers, they have noticed that the RMSE calculated on the first location is **18.52 meters**, this suggests that the true value of the coordinate lies within that range. A factor that could have caused the low accuracy are the presence of trees in the area. On the other hand, the accuracy on the second location is within **6.79 meters**. This is a normal range for a low-cost commercial GPS, however, according to the datasheet the NEO-M8N has an accuracy within 2.5 meters (under optimal conditions). It might be good to note that this was tested in a residential area (more reading should be done to understand the reasons why GPS signals are weakened).
